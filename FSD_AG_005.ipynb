 2: Write a function `checkEligibility` that takes an age as a parameter and returns:


```javascript
// Function to check eligibility
function checkEligibility(age) {
    if (age >= 18) {
        return "Eligible";
    } else {
        return "Not Eligible";
    }
}

// Function calls
console.log(checkEligibility(20));
console.log(checkEligibility(16));
```

 3: Write a function `getDayName` that takes a number (1–7) and returns the corresponding day of the week. Use a `switch` statement.

```javascript
// Function to return the day name
function getDayName(day) {
    switch (day) {
        case 1:
            return "Monday";
        case 2:
            return "Tuesday";
        case 3:
            return "Wednesday";
        case 4:
            return "Thursday";
        case 5:
            return "Friday";
        case 6:
            return "Saturday";
        case 7:
            return "Sunday";
        default:
            return "Invalid day number";
    }
}

// Function calls
console.log(getDayName(1));
console.log(getDayName(5));
console.log(getDayName(7));
console.log(getDayName(9));
```
### Output

```text
Monday
Friday
Sunday
Invalid day number
``

4: Rewrite the `checkEligibility` function from Question 2 using the ternary operator.

The **ternary operator (`? :`)** is a shorthand way of writing an `if-else` statement in JavaScript. It evaluates a condition and returns one value if the condition is true and another value if the condition is false. It makes the code shorter and more readable for simple decision-making.

### JavaScript Code

```javascript
// Function using the ternary operator
function checkEligibility(age) {
    return age >= 18 ? "Eligible" : "Not Eligible";
}

// Function calls
console.log(checkEligibility(20));
console.log(checkEligibility(16));
```

### Output

```text
Eligible
Not Eligible
```

5: Given an array `numbers = [1, 2, 3, 4, 5]`, perform the following:

1. Use a **for loop** to print each number.
2. Use a **while loop** to print each number.
3. Use a **for-in loop** to print each index.
4. Use a **for-of loop** to print each number.


```javascript
const numbers = [1, 2, 3, 4, 5]
console.log("Using for loop:");
for (let i = 0; i < numbers.length; i++) {
    console.log(numbers[i]);
}

console.log("Using while loop:");
let i = 0;
while (i < numbers.length) {
    console.log(numbers[i]);
    i++;
}

console.log("Using for-in loop (Indexes):");
for (let index in numbers) {
    console.log(index);
}

console.log("Using for-of loop (Values):");
for (let number of numbers) {
    console.log(number);
}
```

 6: Explain what functions are in JavaScript and provide an example of a function that adds two numbers.


A **function** in JavaScript is a reusable block of code that performs a specific task. Functions help organize programs into smaller, manageable parts and avoid repeating the same code multiple times. A function can accept **parameters** (inputs), perform operations on them, and return a **result**.

Functions improve code readability, reusability, and maintainability. They can be called whenever needed, making programs more efficient.

### Advantages of Functions

- Promote code reusability.
- Reduce code duplication.
- Improve readability and organization.
- Make debugging and maintenance easier.
- Allow passing data using parameters and returning results.

### JavaScript Code

```javascript
// Function to add two numbers
function add(a, b) {
    return a + b;
}

// Function call
let result = add(10, 20);

// Display the result
console.log("Sum:", result);
```

### Output

```text
Sum: 30
```

 7: Convert the `add` function from Question 6 into an arrow function.

An **arrow function** is a shorter way of writing functions in JavaScript. It was introduced in **ES6 (ECMAScript 2015)** to make function syntax simpler and more concise. Arrow functions are commonly used for short functions and callback functions.

### Advantages of Arrow Functions

- Shorter and cleaner syntax.
- Easy to read and write.
- Useful for callback functions.
- Automatically inherits the `this` value from the surrounding scope.

### JavaScript Code

```javascript
// Arrow function to add two numbers
const add = (a, b) => {
    return a + b;
};

// Function call
let result = add(10, 20);

// Display the result
console.log("Sum:", result);
```

### Output

```text
Sum: 30
```


8: Write a function `greet` that takes a name and a callback function. The `greet` function should call the callback with the name.


A **callback function** is a function that is passed as an argument to another function and is executed after a specific task is completed. Callback functions are commonly used in JavaScript for handling events, asynchronous operations, and code reusability.

In this question, the `greet` function accepts a person's name and a callback function. The callback function is then called with the given name.

### JavaScript Code

```javascript
// Callback function
function displayMessage(name) {
    console.log("Hello, " + name + "!");
}

// Function that accepts a callback
function greet(name, callback) {
    callback(name);
}

// Function call
greet("Rahul", displayMessage);
```

### Output

```text
Hello, Rahul!
```


10: Given the string:

```javascript
const text = " Hello World! ";
```

Perform the following:

1. Trim the whitespace.
2. Convert to uppercase.
3. Check if it includes `"World"`.

JavaScript provides several built-in string methods to manipulate text. The `trim()` method removes leading and trailing whitespace, `toUpperCase()` converts all characters to uppercase, and `includes()` checks whether a specific substring exists within the string.

### JavaScript Code

```javascript
// Original string
const text = " Hello World! ";

// Trim the whitespace
const trimmedText = text.trim();

// Convert to uppercase
const upperText = trimmedText.toUpperCase();

// Check if it includes "World"
const containsWorld = trimmedText.includes("World");

// Display the results
console.log("Trimmed Text:", trimmedText);
console.log("Uppercase Text:", upperText);
console.log("Includes 'World':", containsWorld);
```
